# III. Scraping/Get Data

```python
import time
# Inisialisasi list untuk menyimpan data
price = []
lokasi = []
luas_bangunan = []
luas_tanah = []
sertifikat = []
furnish = []
kamar_tidur = []
kamar_mandi = []
listrik = []
garasi = []

# Inisialisasi WebDriver
# driver = webdriver.Chrome()

# Loop untuk memproses halaman
for pagee in range(1, 60):
    url = f"https://www.brighton.co.id/dijual/rumah/balikpapan?page={pagee}"
    driver.get(url)
    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")
    wrapper_divs = soup.find_all('div', class_='col-12 col-md-6 col-lg-4 py-2 px-0 py-sm-2 px-md-2')

    # Proses setiap listing properti di halaman
    for wrapper in wrapper_divs:
        try:
            link = wrapper.find('a', class_="record-page-visitor")['href']
            driver.get(link)
            html_ind = driver.page_source
            soup_ind = BeautifulSoup(html_ind, "html.parser")

            # Temukan semua elemen div dengan class yang benar
            try:
                harga = soup_ind.find_all('div', class_='wrapper-price detail-property pb-2')[0].get_text()
            except IndexError:
                harga = 'N/A'
            price.append(harga)
            
            try:
                bedroom = soup_ind.find_all('p', class_='small-desc-info line-height-125 text-center pt-1')[2].get_text()
            except IndexError:
                bedroom = 'N/A'
            kamar_tidur.append(bedroom)
            
            try:
                bathroom = soup_ind.find_all('p', class_='small-desc-info line-height-125 text-center pt-1')[3].get_text()
            except IndexError:
                bathroom = 'N/A'
            kamar_mandi.append(bathroom)

            info_divs = soup_ind.find_all('div', class_='wrapper-info-detail')
            time.sleep(1)
            for info_div in info_divs:
                p_tags = info_div.find_all('p')
                for i in range(0, len(p_tags), 2):
                    key = p_tags[i].text.strip() if i < len(p_tags) else 'N/A'
                    value = p_tags[i+1].text.strip() if i+1 < len(p_tags) else 'N/A'

                    # Append values to appropriate lists based on the key
                    if key == 'Lokasi':
                        lokasi.append(value)
                    elif key == 'Luas Bangunan':
                        luas_bangunan.append(value)
                    elif key == 'Luas Tanah':
                        luas_tanah.append(value)
                    elif key == 'Sertifikat':
                        sertifikat.append(value)
                    elif key == 'Furnish':
                        furnish.append(value)
                    elif key == 'Garasi':
                        garasi.append(value)
                    elif key == 'Listrik':
                        listrik.append(value)

            # Ensure all lists have the same length
            max_len = max(len(price), len(lokasi), len(luas_bangunan), len(luas_tanah), len(sertifikat), len(furnish), len(garasi), len(kamar_tidur), len(kamar_mandi), len(listrik))
            while len(price) < max_len:
                price.append('N/A')
            while len(lokasi) < max_len:
                lokasi.append('N/A')
            while len(luas_bangunan) < max_len:
                luas_bangunan.append('N/A')
            while len(luas_tanah) < max_len:
                luas_tanah.append('N/A')
            while len(sertifikat) < max_len:
                sertifikat.append('N/A')
            while len(furnish) < max_len:
                furnish.append('N/A')
            while len(garasi) < max_len:
                garasi.append('N/A')
            while len(kamar_tidur) < max_len:
                kamar_tidur.append('N/A')
            while len(kamar_mandi) < max_len:
                kamar_mandi.append('N/A')
            while len(listrik) < max_len:
                listrik.append('N/A')

        except Exception as e:
            print(f"Error processing property at {link}: {e}")
            continue

# Tutup WebDriver
# driver.quit()

# Buat DataFrame
pages = pd.DataFrame()
pages['harga'] = price
pages['lokasi'] = lokasi
pages['luas_bangunan'] = luas_bangunan
pages['luas_tanah'] = luas_tanah
pages['sertifikat'] = sertifikat
pages['furnish'] = furnish
pages['kamar_tidur'] = kamar_tidur
pages['kamar_mandi'] = kamar_mandi
pages['garasi'] = garasi
pages['listrik'] = listrik

pages
```
