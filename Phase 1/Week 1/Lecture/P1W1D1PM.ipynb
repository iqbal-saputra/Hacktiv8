```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# Definisikan X dan y
y=data[['Overall']]
x=data.drop(['Overall','Name'],axis=1)
x
```

```python
x.shape
```

```python
y.shape
```

```python
# Split train dan test

x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.2, random_state=21)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
# define kolom-kolom numerik dan kategorik

kolom_num=x_train.select_dtypes(exclude=['object']).columns.tolist()
kolom_kat=x_train.select_dtypes(include=['object']).columns.tolist()
kolom_kat
```

```python
x_train.DefensiveWorkRate.unique()
```

```python
x_train[kolom_kat]
```

### Encoding

```python
Enc=OrdinalEncoder(categories=[['Low','Medium','High'],
                   ['Low','Medium','High']])

x_train_kat=Enc.fit_transform(x_train[kolom_kat])
pd.DataFrame(x_train_kat)
```

```python
x_test_kat=Enc.transform(x_test[kolom_kat])
x_test_kat
```

### Scaling

```python
sc=StandardScaler()
x_train_num=sc.fit_transform(x_train[kolom_num])
x_test_num=sc.transform(x_test[kolom_num])
```

### Gabung Data Numerik dan Kategorik

```python
kolom_kat
```

```python
# Train
data_final_train=pd.DataFrame(x_train_num,columns=[kolom_num])
data_final_train[kolom_kat]=x_train_kat
data_final_train
```

```python
data_final_test=pd.DataFrame(x_test_num,columns=[kolom_num])
data_final_test[kolom_kat]=x_test_kat
data_final_test
```

# Modelling

```python
# Proses Training data/pembelajaran mesin thd data
model1=LinearRegression()
model1.fit(data_final_train,y_train)
```

```python
# prediksi menggunakan hasil pembelajaran data
y_train_predict=model1.predict(data_final_train)
y_test_predict=model1.predict(data_final_test)
```

# Evaluasi

```python
act_vs_predict_train=y_train.copy()
act_vs_predict_train['prediksi_overall']=y_train_predict
act_vs_predict_train
```
