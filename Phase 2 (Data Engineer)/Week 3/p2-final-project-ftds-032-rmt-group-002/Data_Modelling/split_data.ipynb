```python
import pandas as pd
```

```python
df = pd.read_csv('../Data_Clean/balikpapan_clean.csv')
```

```python
df
```

```python
df['lokasi'].unique()
```

```python
# Daftar kelurahan untuk setiap kecamatan berdasarkan data unik yang ada
Balikpapan_Timur = ['Balikpapan Timur', 'Lamaru', 'Manggar', 'Batakan']
Balikpapan_Barat = ['Balikpapan Barat', 'Kariangau']
Balikpapan_Utara = ['Balikpapan Utara', 'Batu Ampar', 'Muara Rapak', 'Gn. Samarinda']
Balikpapan_Kota = ['Telaga Sari', 'Klandasan Ilir', 'Klandasan Ulu', 'Prapatan']

# Filter the DataFrame for each kecamatan
df_timur = df[df['lokasi'].isin(Balikpapan_Timur)]
df_barat = df[df['lokasi'].isin(Balikpapan_Barat)]
df_utara = df[df['lokasi'].isin(Balikpapan_Utara)]
df_kota = df[df['lokasi'].isin(Balikpapan_Kota)]

# Display the number of entries in each DataFrame
len(df_timur), len(df_barat), len(df_utara), len(df_kota)
```

```python
# Save each DataFrame to a separate CSV file if needed
df_timur.to_csv('../Data_Clean/balikpapan_timur.csv', index=False)
df_barat.to_csv('../Data_Clean/balikpapan_barat.csv', index=False)
df_utara.to_csv('../Data_Clean/balikpapan_utara.csv', index=False)
df_kota.to_csv('../Data_Clean/balikpapan_kota.csv', index=False)
```
