````python
# Import librariees

from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
````

````python
# buat driver
driver = webdriver.Chrome()
````

````python
# sediakan url
url = 'https://www.bhinneka.com/jual?cari=kulkas&order='
# akses website
driver.get(url)
````

````python
# extract html
html = driver.page_source
````

````python
soup = BeautifulSoup(html, 'html.parser')
````

````python
boxes = soup.find_all('div', {'class':'o_wsale_product_information position-relative d-flex flex-column flex-grow-1 flex-shrink-1'})
````

````python
list_produk = []
list_harga = []
list_cicilan = []
list_rating = []
````

````python
for i in boxes:
    
    try:
        # produk
        nama_produk = soup.find_all('a', {'class':'text-primary text-decoration-none'})
        for i in nama_produk:
            list_produk.append(i.get_text())
    except:
        list_produk.append(None)
        
    try:
        # harga
        harga_produk = soup.find_all('span', {'class':'oe_currency_value'})
        for i in harga_produk:
            list_harga.append(i.get_text())
    except:
        list_harga.append(None)
        
    try:
        # cicilan
        cicilan = soup.find_all('span',{'class':'bmd-installment'})
        for i in cicilan:
            list_cicilan.append(i.get_text())
    except:
        list_cicilan.append(None)
        
    try:
        # rating
        rating_divs = soup.find_all('div', {'class': 'bmd-product-rating'})
        for div in rating_divs:
            # Hitung jumlah bintang
            stars = len(div.find_all('i', {'class': 'ika-star-rating'}))
            # Ambil jumlah ulasan dari elemen span
            reviews = div.find('span', {'class': 'bmd-installment'}).get_text().strip('()')
            list_rating.append((stars, int(reviews)))
    except:
        list_rating.append(None)
        
data = pd.DataFrame({
    'Nama_Produk': list_produk,
    'Harga': list_harga,
    'Cicilan': list_cicilan,
    'Rating': list_rating
})

data.info()
````
