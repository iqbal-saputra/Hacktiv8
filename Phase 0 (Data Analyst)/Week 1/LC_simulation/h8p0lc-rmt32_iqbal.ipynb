```python

'''
Nama: Iqbal
Batch: RMT32
'''

# Program Inventory Management

data = []

# fungsi untuk input data
def input_data():
    nama_barang = input("Masukkan nama barang: ")
    kategori_barang = input("Masukkan kategori barang: ")
    harga_barang = input("Masukkan harga (Rupiah): ")
    data.append([nama_barang, kategori_barang, harga_barang])
    print("Terima kasih, data sudah disimpan")

# fungsi untuk melihat data
def display_data():
    if not data:
        print("Data belum tersedia, silahkan pilih menu 1 untuk isi data")
    else:
        print("No.|Nama Barang|Kategori Barang|Harga")
        for idx, item in enumerate(data, start=1):
            print(f"{idx}|{item[0]}|{item[1]}|{item[2]}")

# program utama
def main():
    while True:
        print("\n==============================")
        print("Program Inventory Management")
        print("==============================")
        print("Menu:")
        print("1. Input Data")
        print("2. Menampilkan data")
        print("\nPilih menu atau keluar: exit")
        
        choice = input("Masukkan pilihan: ")
        
        if choice == "1":
            input_data()
        elif choice == "2":
            display_data()
        elif choice.lower() == "exit":
            print("Terima kasih, sampai jumpa!")
            break
        else:
            print("Pilihan tidak valid, silahkan coba lagi")

# menjalankan program
if __name__ == "__main__":
    main()
```
